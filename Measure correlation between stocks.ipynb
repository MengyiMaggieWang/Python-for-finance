```python
import numpy as np
import pandas as pd
from pandas_datareader import data as wb
import matplotlib.pyplot as plt
```

```python
tickers=['PG','BEI.DE']
sec_data=pd.DataFrame()
for t in tickers:
    sec_data[t]=wb.DataReader(t,data_source='yahoo',start='2016-1-1')['Adj Close']
    
```

```python
sec_data.tail()
```

```python
sec_returns=np.log(sec_data/sec_data.shift(1))
```

```python
sec_returns
```

# Correlation and Covariance

```python
PG_var=sec_returns['PG'].var()
PG_var
```

```python
BEI_var=sec_returns['BEI.DE'].var()
BEI_var
```

```python
PG_var_a=sec_returns['PG'].var()*250
PG_var_a
```

```python
BEI_var_a=sec_returns['BEI.DE'].var()*250
BEI_var_a
```

```python
cov_matrix=sec_returns.cov()   ##pandas.DataFrame.cov() --computes pairwise covariance of columns
cov_matrix
```

```python
cov_matrix_a=sec_returns.cov()*250
cov_matrix_a
```

Explain:the variance of P&G stock is 0.042074,
the variance of BEI.DE is 0.036835
the co-variance of  BEI.DE and P&G is 0.012398

```python
corr_matrix=sec_returns.corr() ###pandas.DataFrame.corr() --computes pairwise correlation of columns
corr_matrix

## we don't need to annualize the cprrelation table since it only shows the relationship between the 2 variable
```

```python
import numpy as np 
import pandas as pd 
from pandas_datareader import data as wb
import matplotlib.pyplot as plt 
```

```python
tickers =['^GSPC','^IXIC','^GDAXI'] ##indices are baskets composed of many stocks ,'^FTSE'???
ind_data= pd.DataFrame()
for t in tickers:
    ind_data[t]=wb.DataReader(t,data_source='yahoo',start='2020-1-1')['Adj Close']
```

```python
ind_data.head()
```

```python
ind_data.tail()
```

```python
(ind_data/ind_data.iloc[0]*100).plot(figsize=(15,8))
plt.show()
```

```python
ind_returns=(ind_data/ind_data.shift(1))-1
```

```python
ind_returns.tail()
```

```python
annual_ind_returns = ind_returns.mean()*250 ## whole year *250
annual_ind_returns
```
